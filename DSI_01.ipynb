```python
laptop=df.groupby('brand')['price'].mean().reset_index()
laptop.brand.unique()
laptop=laptop.sort_values(by='brand',ascending=False)
laptop=laptop.head(10)
plt.xticks(rotation=90)
plt.bar(laptop['brand'],laptop['price'],color="#b25fd0")
plt.title("Top 10 Brand with respect to their Price")
plt.xlabel("Brand name")
plt.ylabel("Price of laptop")
plt.show()
```
